# 14.Find Unique Values From Genre 

```python
df.columns
```

```python
df['genre']
```

```python
list1=[]
for item in df['genre']:
    list1.append(item.split(','))
```

```python
list1
```

```python
one_d=[]
for value in list1:
    for value1 in value:
        one_d.append(value1)
        
one_d
```

```python
uni_list=[]
for uni1 in one_d:
    if uni1 not in uni_list:
        uni_list.append(uni1)
        
uni_list
```
